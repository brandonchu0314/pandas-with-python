```python
import pandas as pd
```

```python
# read and show the dataset from .xlsx file
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales
```

```python
# make a pivot table and add an aggregate function
df_sales.pivot_table(index='Gender', aggfunc='sum')
```

```python
# make a pivot table that shows the total of how much males and females spend in each column
df_sales.pivot_table(index='Gender', values=["Quantity","Total"], aggfunc='sum')
```

```python
# make a pivot table that says how much males and females spend in each category
df_sales.pivot_table(index='Gender', 
                     columns="Product line",
                     values="Total",
                     aggfunc='sum')
```
